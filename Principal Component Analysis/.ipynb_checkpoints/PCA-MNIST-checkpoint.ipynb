# Principal Component Analysis on MNIST DataSet

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# MNIST
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
X = X_train[:100].reshape(-1, 28*28)
y = y_train[:100]

# Visualize
fig, axes = plt.subplots(1, 5, figsize=(10, 5))
for i, ax in enumerate(axes):
    ax.imshow(X[i].reshape(28, 28), cmap='gray')
    ax.set_title(f'Label: {y[i]}')
    ax.axis('off')
plt.show()
```

## Feature Selection

Variance threshold is being used to eliminate features that don't vary much across samples.
This means any pixel (feature) that doesnâ€™t vary significantly across images will be removed.

```python
from sklearn.feature_selection import VarianceThreshold

sel = VarianceThreshold(threshold=(.8 * (1 - .8)))

X_sel = sel.fit_transform(X)

print(f"Original shape: {X.shape}, Reduced shape: {X_sel.shape}")

# Visualize sample images
fig, axes = plt.subplots(1, 5, figsize=(10, 5))
for i, ax in enumerate(axes):
    ax.imshow(X_sel[i].reshape(25, 20), cmap='gray')
    ax.set_title(f'Label: {y[i]}')
    ax.axis('off')
plt.show()
```

## PCA using Scikit-Learn

```python
from sklearn.decomposition import PCA

# Applying PCA to retain 95% of the variance
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X)

print(f"Original shape: {X.shape}, Reduced shape: {X_reduced.shape}")
print(f"Explained variance ratio: {pca.explained_variance_ratio_}")

# Reconstruct images
X_reconstructed = pca.inverse_transform(X_reduced)

# Comparing the images
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i in range(5):
    axes[0, i].imshow(X[i].reshape(28, 28), cmap='gray')
    axes[0, i].set_title("Original")
    axes[0, i].axis('off')
    axes[1, i].imshow(X_reconstructed[i].reshape(28, 28), cmap='gray')
    axes[1, i].set_title("Reconstructed")
    axes[1, i].axis('off')
plt.show()
```

## Visualizing the Latent Space

```python
pca = PCA(n_components=2)

X_full = X_train.reshape(-1, 28*28)
y_full = y_train

X_reduced = pca.fit_transform(X_full)

print(f"Original shape: {X.shape}, Reduced shape: {X_reduced.shape}")

# Step 3: Visualize the reduced data
plt.figure(figsize=(10, 8))
scatter = plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_full, cmap='tab10', alpha=0.7, s=5)
plt.colorbar(scatter, label="Digit Label")
plt.title("MNIST Visualization Using PCA (2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

## Implementing PCA with Eigen Decomposition

resource : https://www.turing.com/kb/guide-to-principal-component-analysis

```python
# Centering the Data
x_mean = np.mean(X, axis=0)
X_centered = X - x_mean

# Covariance Matrix
cov_matrix = np.cov(X_centered.T)

# Eigen Decomposition
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Sorting eigenvalues and eigenvectors (descending order of importance of variance)
sorted_indices = np.argsort(eigenvalues)[::-1]  # ArgSort for indeces as well # Descending Order from highest Variance to Lowest
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Reduce dimensionality
k = 50
W = eigenvectors[:, :k]
X_reduced_eigen = X_centered.dot(W) # X_centered @ W

# Reconstruct images for visualization
X_reconstructed_eigen = X_reduced_eigen.dot(W.T) + x_mean

# original vs reconstructed images (Eigen Decomposition)
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i in range(5):
    axes[0, i].imshow(X[i].reshape(28, 28), cmap='gray')
    axes[0, i].set_title("Original")
    axes[0, i].axis('off')
    axes[1, i].imshow(X_reconstructed_eigen[i].reshape(28, 28), cmap='gray')
    axes[1, i].set_title("Reconstructed")
    axes[1, i].axis('off')
plt.show()
```

## Implementing PCA using Single Vector Decomposition

```python
# Centering the Data
X_centered_svd = X - np.mean(X, axis=0)

# Computing the SVD
U, S, Vt = np.linalg.svd(X_centered_svd, full_matrices=False) # To avoid Calcing COV matrices

# Reduce dimensionality
k = 50
X_reduced_svd = np.dot(X_centered_svd, Vt[:k].T) # U[:, :k] @ np.diag(S[:k])

# Reconstruct images for visualization
X_reconstructed_svd = np.dot(X_reduced_svd, Vt[:k]) + np.mean(X, axis=0) # X_reduced_svd @ Vt[:k, :] + np.mean(X, axis=0)

# Visualize original vs reconstructed images (SVD)
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i in range(5):
    axes[0, i].imshow(X[i].reshape(28, 28), cmap='gray')
    axes[0, i].set_title("Original")
    axes[0, i].axis('off')
    axes[1, i].imshow(X_reconstructed_svd[i].reshape(28, 28), cmap='gray')
    axes[1, i].set_title("Reconstructed")
    axes[1, i].axis('off')
plt.show()
```

## Implementing PCA with Pytorch

```python
import torch

device="cuda"

# Convert data to PyTorch tensor
X_tensor = torch.tensor(X, dtype=torch.float32, device=device)

# Center the data
X_centered_torch = X_tensor - torch.mean(X_tensor, axis=0)

# Perform SVD
U, S, Vt = torch.linalg.svd(X_centered_torch)

# Reduce dimensionality
X_reduced_torch = U[:, :k] @ torch.diag(S[:k])

# Reconstruct images
X_reconstructed_torch = X_reduced_torch @ Vt[:k, :] + torch.mean(X_tensor, axis=0)

# Convert back to CPU for visualization
X_reconstructed_torch_cpu = X_reconstructed_torch.cpu().numpy()

# Visualize original vs reconstructed images (PyTorch)
fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for i in range(5):
    axes[0, i].imshow(X[i].reshape(28, 28), cmap='gray')
    axes[0, i].set_title("Original")
    axes[0, i].axis('off')
    axes[1, i].imshow(X_reconstructed_torch_cpu[i].reshape(28, 28), cmap='gray')
    axes[1, i].set_title("Reconstructed")
    axes[1, i].axis('off')
plt.show()

# Visualizing Eigen Faces
fig, axes = plt.subplots(1, 10, figsize=(15, 5))
for i in range(10):
    axes[i].imshow(Vt[i, :].cpu().numpy().reshape(28, 28), cmap='gray')
    axes[i].set_title(f'PC {i+1}')
    axes[i].axis('off')
plt.show()
```

```python
# Comparing GPU vs CPU compute time
import time

# Time on GPU
start = time.time()
U, S, Vt = torch.linalg.svd(X_centered_torch)
end = time.time()
print(f"GPU Time: {end - start:.4f}s")

# Time on CPU
X_cpu = X_tensor.cpu()
start = time.time()
U, S, Vt = torch.linalg.svd(X_cpu)
end = time.time()
print(f"CPU Time: {end - start:.4f}s")
```

## Approximate PCA with Neural Networks
1-layer linear autoencoder to approximate Principal Component Analysis (PCA) and reduce the dimensionality of the MNIST dataset. 


```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# 1-Layer MLP for PCA Approximation
class LinearAutoencoder(nn.Module):
    def __init__(self, input_dim, reduced_dim):
        super(LinearAutoencoder, self).__init__()
        self.encoder = nn.Linear(input_dim, reduced_dim)
        self.decoder = nn.Linear(reduced_dim, input_dim)
    
    def forward(self, x):
        z = self.encoder(x)  # Encode to lower dimension
        x_reconstructed = self.decoder(z)  # Decode back to original dimension
        return x_reconstructed

# DataSet
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))  # Normalize to [-1, 1]
])

mnist_train = datasets.MNIST(root='./data', train=True, transform=transform, download=True)
train_loader = DataLoader(mnist_train, batch_size=128, shuffle=True)

# Flattening Images
input_dim = 28 * 28
reduced_dim = 10  # Number of principal components (higher dimensions like 30-50 retains more information)

# Model, Loss Function, and Optimizer
model = LinearAutoencoder(input_dim, reduced_dim)
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training the Autoencoder
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

num_epochs = 50
losses = []

for epoch in range(num_epochs):
    epoch_loss = 0
    model.train()  # Ensure model is in training mode
    for images, _ in train_loader:
        images = images.view(-1, input_dim).to(device)  # Flatten images
        
        # Forward pass
        reconstructed = model(images)
        loss = criterion(reconstructed, images)
        
        # Backward pass
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        epoch_loss += loss.item()
    
    avg_loss = epoch_loss / len(train_loader)
    losses.append(avg_loss)
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}")

    model.eval()
    with torch.no_grad():
        for images, _ in test_loader:
            images = images.view(-1, input_dim).to(device)  # Flatten images
            reconstructed = model(images)
            
            # Visualize Original and Reconstructed Images
            images = images.view(-1, 28, 28).cpu()
            reconstructed = reconstructed.view(-1, 28, 28).cpu()
            
            fig, axes = plt.subplots(2, 10, figsize=(15, 3))
            for i in range(10):
                # Original images
                axes[0, i].imshow(images[i], cmap='gray')
                axes[0, i].axis('off')
                # Reconstructed images
                axes[1, i].imshow(reconstructed[i], cmap='gray')
                axes[1, i].axis('off')
            plt.suptitle(f"Original Images (Top) and Reconstructed Images (Bottom). Epoch Number {epoch + 1}")
            plt.show()
            break 

# Training Loss 
plt.plot(losses, label='Training Loss')
plt.title('Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.show()

# Testing the model
test_loader = DataLoader(datasets.MNIST(root='./data', train=False, transform=transform, download=True),
                         batch_size=10, shuffle=True)

model.eval()
with torch.no_grad():
    for images, _ in test_loader:
        images = images.view(-1, input_dim).to(device)  # Flatten images
        reconstructed = model(images)
        
        # Visualize Original and Reconstructed Images
        images = images.view(-1, 28, 28).cpu()
        reconstructed = reconstructed.view(-1, 28, 28).cpu()
        
        fig, axes = plt.subplots(2, 10, figsize=(15, 3))
        for i in range(10):
            # Original images
            axes[0, i].imshow(images[i], cmap='gray')
            axes[0, i].axis('off')
            # Reconstructed images
            axes[1, i].imshow(reconstructed[i], cmap='gray')
            axes[1, i].axis('off')
        plt.suptitle(f"Original Images (Top) and Reconstructed Images (Bottom)")
        plt.show()
        break 
```

## Observations

The model has successfully learned to compress and reconstruct the data, as evidenced by the significant loss reduction.

The loss plateaus at around 0.3. This suggests that the model has converged and is no longer significantly improving with additional training. At this point, the model likely cannot further reduce the reconstruction error without increasing its capacity or changing the architecture.

If better reconstruction quality is desired, increase reduced_dim to retain more information.
